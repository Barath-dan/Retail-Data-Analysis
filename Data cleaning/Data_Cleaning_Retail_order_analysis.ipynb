**DATA CLEANING**

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/project1/orders.csv')
```

```python
df
```

```python
df.info()
```

***Inference:***`There is one missing value in Ship Mode`

**HANDLING MISSING VALUES**

```python
for column in df.columns:
  print('*********************','\n',column,'\n','*********************')
  print(df[column].unique())
```

```python
df = df[~df['Ship Mode'].isin(['Not Available', 'unknown'])]
```

```python
df=df.dropna()
```

```python
df[df['cost price']==0].value_counts().sum()
```

```python
df[df['List Price']==0].value_counts().sum()
```

Both Cost price and List price values have same number of missing **values**

```python
df = df[(df['cost price'] != 0) & (df['List Price'] != 0)]
```

```python
df.info()
```

**FINDING DUPLICATES**

```python
df[df.duplicated()]
```

***Inference:***`There are no duplicate values present in the dataset`

**CONVERTING DATA TYPES**

```python
df['Order Id']=df['Order Id'].astype(str)
```

```python
df['Postal Code']=df['Postal Code'].astype(str)
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],format='%Y-%m-%d')
```

```python
df.info()
```

**RENAMING COLUMNS**

```python
df.columns
```

Changing to lower case

```python
df.columns = [column.lower() for column in df.columns]
```

```python
df.columns
```

Adding underscore '_'

```python
df.columns=[column.replace(' ','_') for column in df.columns]
df.columns
```

Removing Column Country as it has only one value (United States)

```python
df.drop('country', axis=1, inplace=True)
```

**FINAL DATAFRAME AFTER CLEANING**

```python
df
```

```python
df['order_id']=range(1,len(df)+1)
df
```

```python
df.to_csv('/content/project1/Data_Cleaning.csv',index=False)
```
