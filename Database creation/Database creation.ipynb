```python
def insert_data(cursor,connection,df,table):
    try:
        #Creating the insert query dynamically based on the DataFrame's columns
        insert_query = f"INSERT INTO {table_name} ({', '.join(df.columns)}) VALUES ({', '.join(['%s'] * len(df.columns))})"
        #Converting the DataFrame rows to a list of tuples as execute many requires input as list of rows
        data = [tuple(row) for row in df.values]
        #Performing bulk insert using executemany()
        cursor.executemany(insert_query, data)
        connection.commit()
        print(f'Data inserted into {table_name} successfully')

    except connector.Error as err:
        print(f"Connection Error: {err}")
    except Exception as e:
        print(f"An error occurred: {e}")
```
